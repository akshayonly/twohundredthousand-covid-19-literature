```python
sec_source_present = data[data['SI']!='NONE']
```

```python
ct_ids = list()
for ext_source_text in sec_source_present['SI']:
  if 'ClinicalTrials' in ext_source_text:
    source_list = ext_source_text.split(';')
    
    for item in source_list:
      if item.startswith("C"):
        items = item.split('/')
        ct_ids.extend(items[1].split(','))
    
```

```python
genbank_ids = list()
for ext_source_text in sec_source_present['SI']:
  if 'GENBANK' in ext_source_text:
    source_list = ext_source_text.split(';')
    
    for item in source_list:
      if item.startswith("G"):
        items = item.split('/')
        genbank_ids.extend(items[1].split(','))
```
